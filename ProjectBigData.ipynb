```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/egphotelbookings.csv")
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
rating = []
for row in df['arrival_date_month']:
  if row == 'August' :    rating.append(8)
  elif row =='July':   rating.append(7)
  elif row =='May':   rating.append(5)
  elif row =='October':   rating.append(10)
  elif row =='April':   rating.append(4)
  elif row =='June':   rating.append(6)
  elif row =='September':   rating.append(9)
  elif row =='March':   rating.append(3)
  elif row =='February':   rating.append(2)
  elif row =='November':   rating.append(11)
  elif row =='December':   rating.append(12)
  elif row =='January':   rating.append(1)
```

```python
df['Month_number'] = rating
```

```python
df.head()
```

```python
df.shape
```

```python
valid_hotels = ['Renaissance Hotel', 'JW Marriott Hotel']

df = df[df['hotel'].isin(valid_hotels)]

df.shape
```

```python
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(0)
```

```python
df['lead_time'] = df['lead_time'].astype(int)
df.head()
```

```python
# df['adr'] = df['adr'].astype(int)
# df.head()
```

```python
df.drop('arrival_date_week_number', axis=1, inplace=True)
df.head()
```

```python
# Remove the 'company' column
df.drop('company', axis=1, inplace=True)
df.head()
```

```python
df['is_canceled'].isnull().values.any()
```

```python
df['country'].fillna('Unknown', inplace=True)  
df.head()
```

```python
# Create a new column 'guests' by summing the values of 'adults', 'children', and 'babies'
df['guests'] = df['adults'] + df['children'] + df['babies']

df['guests'].fillna(0, inplace=True)  

df['guests'] = df['guests'].astype(int)

df.head()
```

```python
df['total_nights_stayed'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# df.drop(['stays_in_weekend_nights', 'stays_in_week_nights'], axis=1, inplace=True)

df.head()
```

```python
df['meal'] = df['meal'].replace('Undefined', 'SC')

df.head()
```

```python
outlier_cols = ['lead_time', 'adr']
for col in outlier_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
   
```

```python
df['reserved_room_type'].fillna(df['assigned_room_type'], inplace=True)                                 

df['assigned_room_type'].fillna(df['reserved_room_type'], inplace=True)

df.head()

df['coutry'].filla(df['coutry'], inplace=True)
df.head(10)
```

```python
distribution_channels = df['distribution_channel'].unique()
print(distribution_channels)
```

```python
df = df[(df['distribution_channel'] == 'Corporate') | (df['distribution_channel'] == 'Direct') | (df['distribution_channel'] == 'TA/TO') | (df['distribution_channel'] == 'GDS')]
df.head()
```

```python
deposit_types = df['deposit_type'].unique()
print(deposit_types)
```

```python
df = df[(df['deposit_type'] == 'No Deposit') | (df['deposit_type'] == 'Refundable') | (df['deposit_type'] == 'Non Refund')]
df.head()
```

```python
customer_types = df['customer_type'].unique()
print(customer_types)
```

```python
df = df[(df['customer_type'] == 'Transient') | (df['customer_type'] == 'Contract') | (df['customer_type'] == 'Transient-Party') | (df['customer_type'] == 'Group')]
```

```python
df['arrival_date_year'] = df['arrival_date_year'].astype(str)
df['arrival_date_month'] = df['arrival_date_month'].astype(str)
df['arrival_date_day_of_month'] = df['arrival_date_day_of_month'].astype(str)

df['arrival_date'] = pd.to_datetime(df['arrival_date_year'] + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month'],
                                    errors='coerce')

                                                
#df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis=1, inplace=True)

df = df[df['arrival_date'].notnull()]

df.head()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(0)
```

```python
import calendar

monthly_bookings = df.groupby('arrival_date_month')['hotel'].count()

plt.figure(figsize=(10, 6))

month_order = [calendar.month_name[i] for i in range(1, 13)]
monthly_bookings = monthly_bookings.reindex(month_order)
monthly_bookings.plot(kind='line', marker='o')

plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.title('Monthly Booking Trends')
plt.xticks(rotation=45)

plt.show()

```

Monthly Booking Trends where Group the data by month and count the number of bookings. as we can see , from june till august has the highest number of bookings (almost 14000 bookings ) and  from october till january it has the least bookings(7000 bookings and less). it also stars getting a little more bookings in between feb and may but its not the peak

```python
cancel_rate_by_month = df.groupby('arrival_date_month')['is_canceled'].mean()

plt.figure(figsize=(10, 6))
month_order = [calendar.month_name[i] for i in range(1, 13)]
cancel_rate_by_month = cancel_rate_by_month.reindex(month_order)

cancel_rate_by_month.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.title('Cancellation Rate by Month')
plt.xticks(rotation=45)

plt.show()

```
