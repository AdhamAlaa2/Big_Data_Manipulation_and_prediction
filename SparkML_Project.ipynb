```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/egphotelbookings.csv")
```

#Data Cleaning

```python
valid_hotels = ['Renaissance Hotel', 'JW Marriott Hotel']

df = df[df['hotel'].isin(valid_hotels)]

df.head()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(0)
```

```python
df['lead_time'] = df['lead_time'].astype(int)
df.head()
```

```python
df['adr'] = df['adr'].astype(int)
df.head()
```

```python
df.drop('arrival_date_week_number', axis=1, inplace=True)
df.head()
```

```python
df.drop('company', axis=1, inplace=True)
df.head()
```

```python
df['is_canceled'].fillna(0, inplace=True)
df.head()

```

```python
df['country'].fillna('Unknown', inplace=True)  
df.head()
```

```python
df['guests'] = df['adults'] + df['children'] + df['babies']

# df.drop(['adults', 'children', 'babies'], axis=1, inplace=True)

df['guests'] = df['guests'].astype(int)

df.head()
```

```python
#df=df[(df['adults'] <= 4) & (df['babies'] <= 4) & (df['children'] <= 4)]
```

```python
df['total_nights_stayed'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df.drop(['stays_in_weekend_nights', 'stays_in_week_nights'], axis=1, inplace=True)

df.head()
```

```python
outlier_cols = ['lead_time', 'adr']
for col in outlier_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
   


```

```python
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

df['arrival_date_month'] = df['arrival_date_month'].map(month_mapping)

df.head()
```

```python
df['meal'] = df['meal'].replace('Undefined', 'SC')

df.head()
```

```python
deposit_types = df['deposit_type'].unique()
print(deposit_types)
```

```python
df = df[(df['deposit_type'] == 'No Deposit') | (df['deposit_type'] == 'Refundable') | (df['deposit_type'] == 'Non Refund')]
df.head()
```

```python
customer_types = df['customer_type'].unique()
print(customer_types)
```

```python
df = df[(df['customer_type'] == 'Transient') | (df['customer_type'] == 'Contract') | (df['customer_type'] == 'Transient-Party') | (df['customer_type'] == 'Group')]
df.head()
```

```python
distribution_channels = df['distribution_channel'].unique()
print(distribution_channels)
```

```python
df = df[(df['distribution_channel'] == 'Corporate') | (df['distribution_channel'] == 'Direct') | (df['distribution_channel'] == 'TA/TO') | (df['distribution_channel'] == 'GDS')]
df.head()
```

```python
df['reserved_room_type'].fillna(df['assigned_room_type'], inplace=True)

df['assigned_room_type'].fillna(df['reserved_room_type'], inplace=True)

df.head()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(0)
```

```python
df.to_csv('hotel_dataset_cleaned.csv', index=False)
```

```python
from pyspark.sql import SparkSession
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
spark = SparkSession.builder.appName('assignment').getOrCreate()
cleaned = spark.read.csv('/content/hotel_dataset_cleaned.csv', header=True, inferSchema=True)
```

```python
cleaned.printSchema()
```

```python
cleaned.show(5)
```

```python
stages=[]

from pyspark.sql import SparkSession
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler,StringIndexer,OneHotEncoder
from pyspark.ml.classification import LogisticRegression
from sklearn.metrics import accuracy_score

categorical_columns = ['hotel','meal', 'country', 'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type']

numerical_columns=['lead_time','arrival_date_year','arrival_date_month','arrival_date_day_of_month','adults','children','babies','is_repeated_guest','previous_cancellations','previous_bookings_not_canceled'
                   ,'booking_changes','agent','days_in_waiting_list','adr','required_car_parking_spaces','total_of_special_requests','guests','total_nights_stayed']
indexers = [StringIndexer(inputCol=column, outputCol=column+"_index", handleInvalid="keep") for column in categorical_columns]

indexed_df = cleaned
for indexer in indexers:
    indexed_df = indexer.fit(indexed_df).transform(indexed_df)

indexed_columns=["hotel_index","meal_index","country_index","distribution_channel_index","reserved_room_type_index","assigned_room_type_index","deposit_type_index","customer_type_index"]
output_encoded_columns=[
    "hotel_encoded","meal_encoded","country_encoded","distribution_channel_encoded","reserved_room_type_encoded","assigned_room_type_encoded","deposit_type_encoded","customer_type_encoded"
]

onehot_encoder = OneHotEncoder(inputCols=indexed_columns, outputCols=output_encoded_columns)
stages+=[onehot_encoder]


selected_columns = numerical_columns + [column+"_encoded" for column in categorical_columns]

# Create the VectorAssembler
assembler = VectorAssembler(inputCols=selected_columns, outputCol='features')

stages +=[assembler]



pipeline=Pipeline(stages=stages)
pipeline_model=pipeline.fit(indexed_df)
indexed_df=pipeline_model.transform(indexed_df)


```

```python
print(stages)
```

```python
indexed_df.show(10)
```

```python
train, test = indexed_df.randomSplit([0.8, 0.2], seed = 2018)
```

```python
train.show()
```

RANDOM FOREST MODEL

```python
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator

rf = RandomForestClassifier(featuresCol='features', labelCol='is_canceled', numTrees=10)
rfModel = rf.fit(train)
predictions = rfModel.transform(test)



true_labels=predictions.select('is_canceled')
lr_predictions=predictions.select('prediction')



```

```python
predictions.show()
```

```python

accuracy = accuracy_score(true_labels.toPandas(), lr_predictions.toPandas())
print("Random Forest Accuracy =",accuracy*100,"%")
```

```python

from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(labelCol='is_canceled', metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
print("Random Forest- Accuracy: {:.2%}".format(accuracy))
```

Gradiant Boost classifier

```python
gbt = GBTClassifier(featuresCol='features', labelCol='is_canceled', maxIter=10) 

gbModel = gbt.fit(train)
predictions_gbt = gbModel.transform(test)

true_labels_GBT =predictions_gbt.select('is_canceled')
lr_predictions_GBT =predictions_gbt.select('prediction')

```

```python
accuracy = accuracy_score(true_labels_GBT.toPandas(), lr_predictions_GBT.toPandas())
print("Random Forest Accuracy =",accuracy*100,"%")


```

logistic regression

```python

lr = LogisticRegression(featuresCol='features', labelCol='is_canceled', maxIter=10)

lrModel_logistic_regression = lr.fit(train)
predictions_logistic_regression = lrModel_logistic_regression.transform(test)

true_labels_logistic_regression=predictions_logistic_regression.select('is_canceled')
lr_predictions_logistic_regression =predictions_logistic_regression.select('prediction')
```

```python
accuracy = accuracy_score(true_labels_logistic_regression.toPandas(), lr_predictions_logistic_regression.toPandas())
print("Logistic Regression Accuracy =",accuracy*100,"%")
```
